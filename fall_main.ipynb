```python
#Fall Detection + Face Blurring + Facial Expression Recognition
from platform import python_version
print(python_version())
#Setup
# !pip install -q mediapipe opencv-python deepface
# !pip install tf-keras

#Imports

import mediapipe as mp
from deepface import DeepFace
import numpy as np
import cv2
import math
```

```python
#Initialize MediaPipe Pose
draw = mp.solutions.drawing_utils
pose = mp.solutions.pose.Pose()
```

```python
#Load OpenCV Face Detector
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
#Load Video or Webcam
#cap = cv2.VideoCapture("fall_sample_video.mp4")  # Use 0 for webcam
cap = cv2.VideoCapture(0) 
```

```python
#Saving Output Video
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output_fall_privacy.avi', fourcc, 20.0, (640, 480))
```

```python
#Main loop
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("No frame captured.")
        break

    frame = cv2.resize(frame, (640, 480))
    image_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose.process(image_rgb)

    fall_detected = False
    angle = None

  

    #Detecting body landmarks and calculate angle
    if results.pose_landmarks:
        landmarks = results.pose_landmarks.landmark

        try:
            left_shoulder = landmarks[11]
            right_shoulder = landmarks[12]
            left_hip = landmarks[23]
            right_hip = landmarks[24]

            mid_shoulder_x = (left_shoulder.x + right_shoulder.x) / 2
            mid_shoulder_y = (left_shoulder.y + right_shoulder.y) / 2
            mid_hip_x = (left_hip.x + right_hip.x) / 2
            mid_hip_y = (left_hip.y + right_hip.y) / 2

            dx = mid_hip_x - mid_shoulder_x
            dy = mid_hip_y - mid_shoulder_y

            angle = abs(math.degrees(math.atan2(dy, dx)))

            # Fall condition: body is close to horizontal
            if angle < 50 or angle > 130:
                fall_detected = True

            draw.draw_landmarks(frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

        except Exception as e:
            print("Pose processing error:", e)

    # Detecting and blur face, and detect emotion
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    for (x, y, w, h) in faces:
        face_roi = frame[y:y+h, x:x+w]

        #Facial expression recognition
        try:
            result = DeepFace.analyze(face_roi, actions=['emotion'], enforce_detection=False)
            expression_label = result[0]['dominant_emotion']
            cv2.putText(frame, expression_label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,0), 2)
        except:
            expression_label = "Unknown"

        # Blur the face for privacy
        ksize = (w//2*2+1, h//2*2+1)  # must be odd numbers
        blur = cv2.GaussianBlur(face_roi, ksize, 30)
        frame[y:y+h, x:x+w] = blur
        # blur = cv2.GaussianBlur(face_roi, (99, 99), 30)
        # frame[y:y+h, x:x+w] = blur

    #Annotating output
    if fall_detected:
        cv2.putText(frame, "FALL DETECTED!", (20, 60), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 0, 255), 3)

    if angle is not None:
        cv2.putText(frame, f"Body Angle: {int(angle)}", (20, 100), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255,255,0), 2)

    #Show and save frame
    cv2.imshow("Fall Detection System", frame)
    out.write(frame)

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

# ðŸ”š Cleanup
cap.release()
out.release()
cv2.destroyAllWindows()
```
